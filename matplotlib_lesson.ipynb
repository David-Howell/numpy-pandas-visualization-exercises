## Example 1

```python
from random import randint

x1 = list(range(10))
y1 = [randint(0,10) + j for j in x1]

x2 = list(range(10))
y2 = [randint(0,10) + j for j in x2]

x3 = list(range(10))
y3 = [randint(0,10) + k for k in x3]
```

### Basic plot

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.show
```

### Custom ticks & rotate

- .xticks()
- .xticks(rotation=)

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)

plt.xticks([0, 2, 4, 6, 8],
          ['zero', 'two', 'four', 'six', 'eight'],
          rotation = 45)


plt.show
```

### Markers & line styles

- plot(marker=, ls=)

```python
plt.plot(x1,y1, marker='o', ls='--')
plt.plot(x2,y2)
plt.plot(x3,y3)

plt.xticks([0, 2, 4, 6, 8],
          ['zero', 'two', 'four', 'six', 'eight'],
          rotation = 45)


plt.show
```

### Figure & Subplots

- .figure(figsize=()), .subplot()

```python
plt.figure(figsize=(12,6))

plt.plot(x1,y1, marker='o', ls='--')
plt.plot(x2,y2)
plt.plot(x3,y3)

plt.xticks([0, 2, 4, 6, 8],
          ['zero', 'two', 'four', 'six', 'eight'],
          rotation = 45)


plt.show
```

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.plot(x1,y1, marker='o', ls='--')

plt.subplot(132)
plt.plot(x2,y2)

plt.subplot(133)
plt.plot(x3,y3)

plt.xticks([0, 2, 4, 6, 8],
          ['zero', 'two', 'four', 'six', 'eight'],
          rotation = 45)


plt.show
```

### Add limits

- .xlim(), .ylim()

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.plot(x1,y1, marker='o', ls='--')
plt.ylim(0,18)

plt.subplot(132)
plt.plot(x2,y2)
plt.ylim(0,18)

plt.subplot(133)
plt.plot(x3,y3)
plt.ylim(0,18)

plt.xticks([0, 2, 4, 6, 8],
          ['zero', 'two', 'four', 'six', 'eight'],
          rotation = 45)


plt.show
```

### Add text, lines, & arrows!!

> .text, .hlines, .vlines, .annotate

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.plot(x1,y1, marker='o', ls='--')
plt.ylim(0,18)
plt.hlines(10,0,10, ls='--')

plt.subplot(132)
plt.plot(x2,y2)
plt.ylim(0,18)
plt.text(1,11.2, '(1,11)')

plt.subplot(133)
plt.plot(x3,y3)
plt.ylim(0,18)
plt.annotate('look here',
            xy = (5,14),)

plt.xticks([0, 2, 4, 6, 8],
          ['zero', 'two', 'four', 'six', 'eight'],
          rotation = 45)


plt.show
```

### Basic plot

- bmi vs charges

```python
plt.scatter(df.bmi,df.charges)
plt.show
```

### Basic plot - two sets of data

- bmi vs charges for smoker and non-smoker

```python
df_smoke_no = df [df.smoker == 'no']
df_smoke_yes = df [df.smoker == 'yes']
```
