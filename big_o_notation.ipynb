Curves to graph

$y=0n+1$ and label the curve "O(1)"

$y=log(n)$ and label the curve "O(log n)"

$y=n$ and label the curve "O(n)"

$y=n∗log(n)$ and label it "O(n log n)"

$y=n^2$ and label it "O(n^2)"

$y=2^n$ and label it "O(2^n)"

$y=n!$ and label it "O(n!)"

$y=n^n$ and label it "O(n^n)"

```python
t1 = '$O(1)$ '
t2 = '$O(log n)$ '
t3 = '$O(n)$ '
t4 = '$O(n log n)$'
t5 = '$O(n^2)$'
t6 = '$O(2^n)$'
t7 = '$O(n!)$' 
t8 = '$O(n^n)$' 
```

```python
# t1 = '$y=0n+1$ '
# t2 = '$y=log(n)$ '
# t3 = '$y=n$ '
# t4 = '$y=n∗log(n)$'
# t5 = '$y=n^2$'
# t6 = '$y=2^n$'
# t7 = '$y=n!$' 
# t8 = '$y=n^n$' 
```

```python
x = [] 
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []


for n in range(1,11):
    x += [n]
    y1 += [0*n + 1]
    y2 += [m.log(n)]
    y3 += [n]
    y4 += [n * m.log(n)]
    y5 += [n * n]
    y6 += [2 ** n]
    y7 += [m.factorial(n)]
    y8 += [n ** n]
```

```python
plt.figure(figsize=(12,10))

plt.plot(x,y1, color = 'red', label = t1)

plt.plot(x,y2, color = 'darkorange', label = t2)

plt.plot(x,y3, color = 'yellow', label = t3)

plt.plot(x,y4, color = 'green', label = t4)

plt.plot(x,y5, color = 'blue', label = t5)

plt.plot(x,y6, color = 'indigo', label = t6)

plt.plot(x,y7, color = 'violet', label = t7)

plt.plot(x,y8, color = 'cyan', label = t8)

plt.xlim(1,10)
plt.ylim(-1,y8[3])
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc=2, fontsize=13)

plt.show
```
