1. Use matplotlib to plot the following equation:

      ### $y = x^2 âˆ’ x +2$

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = [n for n in range(-30,31)]
y = [n**2 - n + 2 for n in range(-30,31)]
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y, c='red')
plt.vlines(0, -50, 1000, color='black')
plt.hlines(0, -30, 30, color='black')
plt.xlim(-30,30)
plt.ylim(-30, 1000)
plt.annotate('The Origin',
            xy = (0,0),
            xytext = (4,200),
            rotation = 30,
            arrowprops={'facecolor': 'blue'})
plt.show
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

$y=âˆšx$

$y=x^3$

$y=2^x$

$y=1/(x+1)$


You can use functions from the `math` module to help implement some of the equations above.

```python
t1 = '$y=âˆšx$'
t2 = '$y=x^3$'
t3 = '$y=2^x$'
t4 = '$y=1/(x+1)$'
```

```python
import math
```

```python
x1 = []
y1 = []
x2 = []
y2 = []
x3 = []
y3 = []
x4 = []
y4 = []

for x in range(1,201):
    x1 += [x]
    y1 += [math.sqrt(x)]
    

for x in range(-10,11):
    x2 += [x]
    y2 += [x**3]
    
for x in range(-10,11):
    x3 += [x]
    y3 += [2**x]
    
for x in range(-100,101):
    x /= 10
    if x == -1:
        continue
    x4 += [x]
    y4 += [1/(x+1)]
```

```python
plt.figure(figsize=(12,6))

plt.plot(x1,y1)
plt.title(t1)
plt.hlines(0, -10, x1[-1], color='black')
plt.vlines(0, -1, y1[-1], color='black')

plt.show
```

```python
plt.figure(figsize=(12,6))

plt.plot(x2,y2)
plt.title(t2)
plt.hlines(0, -11, 11, color='black')
plt.vlines(0, -1010, 1010, color='black')

plt.show
```

```python
plt.figure(figsize=(12,6))

plt.plot(x3,y3)
plt.title(t3, fontsize=20)
plt.hlines(0, -10, 10, color='black')
plt.vlines(0, 0, 100, color='black')
plt.xlim(-5,10)
plt.ylim(0,100)

plt.show
```
